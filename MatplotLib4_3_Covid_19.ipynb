```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
web_page = pd.read_html('https://en.wikipedia.org/wiki/COVID-19_pandemic_in_India',"Recoveries")[1]
```

```
web_page
```

```
web_page.iloc[0:36,0] # 0 se 35 rows tk chahiye column 0 ki
```

```
web_page = web_page.iloc[:,:5]
```

```
web_page
```

```
web_page.columns = ['states','total cases','death','recoveries','active cases'] #modify column headers
```

```
web_page
```

```

web_page.drop([36,37],inplace=True)
```

```
web_page
```

```
web_page['total cases'][3]=web_page['total cases'][3].strip('[]bc')#clean data from this special characters and alphabets
web_page['total cases'][3]=web_page['total cases'][3].replace(',','')
web_page['death'][16]=web_page['death'][16].strip('[]bc,')#because we convert into integer
```

```
web_page
```

```
# Now convert all columns values into integer
TC=pd.Series(map(int,web_page['total cases']))
DH=pd.Series(map(int,web_page['death']))
REC=pd.Series(map(int,web_page['recoveries']))
AC=pd.Series(map(int,web_page['active cases']))
```

```
fig= plt.figure(figsize=(15,6))
axes=fig.add_axes([0,0,1,1])
x=np.arange(0,36*4,step=4) # 4 bar bnae h is liye 36*4 lra h or steps bhi 4 rkhe h
axes.bar(x,TC,width=0.8,label='Total cases',color='c')
axes.bar(x+0.8,REC,width=0.8,label='Recoveries') #shift x 0.4to the previous bar
axes.bar(x+1.6,AC,width=0.8,label='Active Cases')
axes.bar(x+2.4,DH,width=0.8,label='Death')
plt.xticks(x+1.6,web_page['states'],rotation=90) # numbers ko centre m kra h or numbers ko states se replace kra h or
# 90 degre se rotate bhi kra h q k nam overlap horhe the
axes.legend() # lebels show krne k liye legend use krte h
plt.show()
```
